```python
# Show the first 10 rows in dataframe
import pandas as pd
df = pd.read_csv("https://data.buffalony.gov/resource/n4ni-uuec.csv")
df
```

```python
import numpy as np
# Replace the values "VACANT" and 0 with NaN
df = df.replace("VACANT", np.nan)
df = df.replace(0, np.nan)
df = df.replace("0", np.nan)
df
```

```python
# Calculate the number of missing values in each column
missing_value_counts = df.isna().sum()
missing_value_counts
```

```python
# sort the missing_value_counts from most missing value in columns to least
missing_value_counts.sort_values(ascending=False, inplace=True)
# Get the header and convert to list
most_missing_value_counts = list(missing_value_counts.head(4).index)
most_missing_value_counts
```

```python
# Delete the column named in most_missing_value_counts list 
for i in most_missing_value_counts:
  df.drop([i], axis=1, inplace=True)
df
```

```python
df.isna().sum()
```

```python
# Find the number of missing values in each row
missing_values_count = df.isna().sum(axis=1)
# Get the rows with missing values
rows_with_missing_values = missing_values_count[missing_values_count > 0]
rows_with_missing_values = list(rows_with_missing_values.index)
print(rows_with_missing_values)
```

```python
# Delete the rows with missing values from the DataFrame
df.drop(rows_with_missing_values, axis=0, inplace=True)
df
```

```python
# Reset the index of the DataFrame
df.reset_index(inplace=True)

# Check the number of missing values in each row
missing_values_count = df.isna().sum(axis=1)
missing_values_count
```

```python
df
```

```python
# The target variable is Common Name
target = "common_name"

# Load the Boston Housing dataset
X = df.drop([target,"index", "geocoded_column","editing", "street", "side", "latitude",	"longitude", "site_id"], axis = 1)
y = df[target]
X
```

```python
y
```

```python
# Find all the categorical string columns in the dataframe
categorical_cols = X.select_dtypes(include='object').columns
categorical_cols
```

```python
# Convert the categorical string columns to numerical data using get_dummies
df_numerical = pd.get_dummies(X, columns=categorical_cols)
df_numerical
```

```python
# Combine the numerical columns with the original numerical columns
df_final = pd.concat([df_numerical, X[X.columns.difference(categorical_cols)]], axis=1)
df_final
```

```python
df_final.columns
```

```python
from sklearn.model_selection import train_test_split


X = df_final

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

display(X_train)
display(X_test)
display(y_train) 
display(y_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train the classifier
lr = LogisticRegression(multi_class='ovr', solver='lbfgs')
lr.fit(X_train_scaled, y_train)

# Make predictions on the testing set
y_pred = lr.predict(X_test_scaled)

# Evaluate the classifier's accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Test Accuracy is: {:.2f}% for Logistic Regression".format(accuracy * 100))
```

```python
from sklearn import svm

# Train the classifier
svc = svm.SVC(kernel='linear')
svc.fit(X_train_scaled, y_train)

# Make predictions on the testing set
y_pred = svc.predict(X_test_scaled)

# Evaluate the classifier's accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Test Accuracy is: {:.2f}% for SVM".format(accuracy * 100))
```

```python
from sklearn import tree

# Train the classifier
dtc = tree.DecisionTreeClassifier()
dtc.fit(X_train_scaled, y_train)

# Make predictions on the testing set
y_pred = dtc.predict(X_test_scaled)

# Evaluate the classifier's accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Test Accuracy is: {:.2f}% for Decision Tree Classifier".format(accuracy * 100))
```
